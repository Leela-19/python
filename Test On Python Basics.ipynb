```python
#  6 Write a Python program to check if it is a strong number or not


import math
a='145'
print(f'given number is {int(a)}')
b=[]
d=0
for i in a:
  c=math.factorial(int(i))
  b.append(c)
print(f'factorials of each digit are : {b}')

for j in b:
  d=d+j
print(f'sum of factorials is : {d}')
if int(a)==d:
  print(f'{a} is a strong number')
else:
  print(f'{a} is not a strong number')
```
